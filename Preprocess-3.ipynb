```python
import numpy as np
import pandas as pd
import pickle as pkl
import matplotlib as plt
import os
```

```python
with open('pickles\\Removed_Others_Dialogue.pkl', 'rb') as fl:
    df = pkl.load(fl)
df
```

```python
df.reset_index(inplace=True, drop=True)
df
```

```python
df.transcript.nunique()
```

```python
import re

def clean_transcript(single_str):
    lt_names= ['PRESIDENT RODRIGO DUTERTE:', 'PRESIDENT DUTERTE:', 'PRESIDENT RODRIGO ROA DUTERTE:', 'DUTERTE:', 
               'PRES. DUTERTE:', 'PRRD']
    str_cleaned = single_str
    for nm in lt_names:
        str_cleaned = str_cleaned.replace(nm, ' ')
    str_cleaned = str_cleaned.replace('\xa0', ' ')
    str_cleaned = str_cleaned.replace('\x0c', ' ')
    str_cleaned = str_cleaned.replace('\\', ' ')
    str_cleaned = str_cleaned.replace('—', ' ')
    str_cleaned = str_cleaned.replace('-', ' ')
    str_cleaned = str_cleaned.replace('...', ' ')
    str_cleaned = str_cleaned.replace('…', ' ')
    str_cleaned = str_cleaned.replace('END', ' ')
    str_cleaned = re.sub(r"\[(.*?)\]", ' ', str_cleaned) #remove all that is contained within brackets
    str_cleaned = re.sub(r"\((.*?)\)", ' ', str_cleaned) #remove all that is contained within parentheses
    str_cleaned = re.sub(r'\“(.+?)\”', ' ', str_cleaned) #remove all that is in quatation as he probably quoting someone
    str_cleaned = str_cleaned.replace('‘', ' ')
    str_cleaned = [wrd.lower() for wrd in str_cleaned.split(' ') if wrd != '']
    str_cleaned = ' '.join(str_cleaned)

    return str_cleaned
```

>We're going to apply the function above to every transcript for cleaning. But before that let's correct the indexing

```python
df['transcript'] = df.transcript.map(clean_transcript)
df
```

>Let's parse date features/variables/columns into date object

```python
df['date'] = df.date.map(lambda x: x if x.find('_') == -1 else x[:-2]).tolist()
df['date'] = pd.to_datetime(df['date'])
df
```

>Let's group these records by transcripts(since there are duplciates)

```python
grouped_transcripts = df.groupby('transcript')
```

>Let's find out whose group contains multiple record-basically a duplicate.

```python
multiples = np.array([])

for i in grouped_transcripts.groups:
    #if it contain more than 1 record then it has duplicate
    if len(grouped_transcripts.get_group(i)) > 1:
        #append the group to multiples
        multiples = np.append(multiples, [i])
len(multiples)
```

>So there are 8 transcripts that have duplicates. I checked them individually and they really are duplicates so I'm going to remove the duplicates.

```python
df_temp = df.copy()
```

```python
for group in multiples:
    df.drop(index=grouped_transcripts.get_group(group).index.values[1:], inplace=True)
    grouped_transcripts = df.groupby('transcript')
```

```python
df.reset_index(inplace=True, drop=True)
print('Number of unique transcripts:', df.transcript.nunique())
print('Number of transcripts in our data:', df.shape[0])
```
