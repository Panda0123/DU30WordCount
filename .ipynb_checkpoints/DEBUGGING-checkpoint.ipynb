```python
full_txt = ''
isDuterte=True
lt_names= ['PRESIDENT RODRIGO DUTERTE:', 'PRESIDENT DUTERTE:', 'PRESIDENT RODRIGO ROA DUTERTE:', 'DUTERTE:', 
               'PRES. DUTERTE:', 'PRRD']
for i in range(3, len(p_list)):
    p = p_list[i]
    if p.strong:
        if any(p.strong.text.strip() == nm for nm in lt_names):
            isDuterte = True
            full_txt += '\n'
        else:
            isDuterte = False
    
    if isDuterte:
        full_txt += p_list[i].text
```

```python
url = 'https://www.mindanews.com/duterte-files/2018/02/duterte-transcripts-ip-leaders-summit-as-delivered-01-feb-2018__trashed/'
url = 'https://www.mindanews.com/duterte-files/2017/10/duterte-transcripts-ibp-davao-30-sept-2017__trashed/'
response = requests.get(url, headers=headers)
html = BeautifulSoup(response.text, 'lxml')
```

```python
import re
container = html.find('div', 'td-post-content')
p_list = container.findAll('p')
pattern_index = 0
span = [0, 1]
for i in range(len(p_list)):
    loc = p_list[i].text.strip()
    if re.search(r'\[(.*?)\]', loc):
        pattern_index = i
        span = re.search(r'\[(.*?)\]', loc).span()
        break
pattern_str = loc[span[0]:span[1]]
loc = pattern_str[pattern_str.find('|')+ 1:-1].strip()
date = pattern_str[1:pattern_str.find('|')].strip()

#some are formated differently
if re.search('(\d){1,2} \w+ \d{4}', loc):
    loc = pattern_str[1:pattern_str.find('|')].strip()
    date = pattern_str[pattern_str.find('|') + 1:-1].strip()

if bool(re.search('(\d){1,2} \w+ \d{4}', date)):
    date = format_date(date)
```
