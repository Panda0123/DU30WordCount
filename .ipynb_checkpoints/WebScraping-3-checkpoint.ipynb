```python
def format_date(dt):
    #format properly
    day = dt[:2]
    month_yr = dt[2:].strip()
    dt = f'{month_yr[:-4].strip()} {day}, {month_yr[-4:]}' 
    return dt
```

>Let's get first the information presented in the duterte files browser. 

```python
all_info = {}
dates = {}
headers = {
    'user-agent': 'Mozilla/5.0 (iPhone; CPU iPhone OS 10_3 like Mac OS X) AppleWebKit/602.1.50 (KHTML, like Gecko) CriOS/56.0.2924.75 Mobile/14E5239e Safari/602.1 RuxitSynthetic/1.0 v1355348020 t3296535826494656701 smf=0'
}

#apparently only the first 7 pages are available all other else respond with 404
for i in range(7):
    url = f'https://www.mindanews.com/category/duterte-files/page/{i+1}/'
    response = requests.get(url, headers=headers)
    html = BeautifulSoup(response.text, 'lxml')
    
    container = html.find('div', class_='td-ss-main-content')
    div_list = container.findAll('div', 'item-details')
    for div in div_list:
        header = div.find('h3').text
        transcripts_headers_offset = ['DUTERTE TRANSCRIPTS', 'DUTERTE TRANCRIPTS', 'DIGONG TRANSCRIPTS']
        if any(header[:header.find(':')].strip() == nm or \
           header[:header.find('.')].strip() == nm \
           for nm in transcripts_headers_offset):
            
            #some does not contain dates
            if header.rfind('.') != -1:
                if header.find(':') != -1:
                    title = header[header.find(':')+1:header.rfind('.')].strip()
                else:
                    title = header[header.find('.')+1:header.rfind('.')].strip()
                    
                lnk = div.find('h3').a['href'].strip()
                date = format_date(header[header.rfind('.')+ 1:].strip())
            else:
                title = header[header.find(':')+1:].strip()
                lnk = div.find('h3').a['href'].strip()

            if date in dates:
                all_info[f'{date}_{dates[date]+1}'] = {'date':date, 'title':title, 'url':lnk}
                dates[date] += 1
            else:
                all_info[date] = {'date':date, 'title':title, 'url':lnk}
                dates[date] = 1
        else:
            print('!', header)
```

```python
with open('pickles\\dictionary_wo_trans.pkl' , 'wb') as fl:
    pkl.dump(all_info, fl)
```

>Now we're going to extract the transcript 

```python
import re
def multi_dialogue(html):
    container = html.find('div', 'td-post-content')
    p_list = container.findAll('p')
    #keep track when pattern is found
    pattern_index = 0
    span = [0, 1]
    for i in range(len(p_list)):
        loc = p_list[i].text.strip()
        if re.search(r'\[(.*?)\]', loc):
            pattern_index = i
            span = re.search(r'\[(.*?)\]', loc).span()
            break
    
        
    pattern_str = loc[span[0]:span[1]]
    loc = pattern_str[pattern_str.find('|')+ 1:-1].strip()
    
    #some are formated differently
    if re.search('(\d){1,2} \w+ \d{4}', loc):
         loc = pattern_str[1:pattern_str.find('|')].strip()
    
    if len(loc) < 6:
        loc = pattern_str[1:-1]
        
    full_txt = ''
    isDuterte=True
    lt_names= ['PRESIDENT RODRIGO DUTERTE:', 'PRESIDENT DUTERTE:', 'PRESIDENT RODRIGO ROA DUTERTE:', 'DUTERTE:', 
                   'PRES. DUTERTE:', 'PRRD']
    #we're going to start below the pattern
    for i in range(pattern_index+1, len(p_list)):
        p = p_list[i]
        if p.strong:
            if any(p.strong.text.strip() == nm for nm in lt_names):
                isDuterte = True
                full_txt += '\n'
            else:
                isDuterte = False

        if isDuterte:
            full_txt += p_list[i].text
    return full_txt, loc

def ordinary(html):
    container = html.find('div', 'td-post-content')
    p_list = container.findAll('p')
    pattern_index = 0
    span = [0, 1]
    for i in range(len(p_list)):
        loc = p_list[i].text.strip()
        if re.search(r'\[(.*?)\]', loc):
            pattern_index = i
            span = re.search(r'\[(.*?)\]', loc).span()
            break
    
        
    pattern_str = loc[span[0]:span[1]].strip()
    loc = pattern_str[1:pattern_str.find('|')].strip()
    #some are formated differently
    if re.search('(\d){1,2} \w+ \d{4}', loc):
         loc = pattern_str[1:pattern_str.find('|')].strip()
    
    if len(loc) < 6:
        loc = pattern_str[1:-1]
    
    if 'Delivered at' in loc:
        loc = loc[loc.find('Delivered at')+12:].strip()
    full_txt = ''
    for i in range(pattern_index+1, len(p_list)):
        full_txt += p_list[i].text.strip()
    return full_txt, loc
```

```python
trans = []
locations = []
for key in all_info.keys():
    try:
        url = all_info[key]['url']
        response = requests.get(url, headers=headers)
        html = BeautifulSoup(response.text, 'lxml')
        lt_names= ['PRESIDENT RODRIGO DUTERTE:', 'PRESIDENT DUTERTE:', 'PRESIDENT RODRIGO ROA DUTERTE:', 'DUTERTE:', 
                      'PRES. DUTERTE:', 'PRRD']
        
        full_txt, loc = ordinary(html)
        
        if any(nm in full_txt for nm in lt_names):
            full_txt, loc = multi_dialogue(html)

        trans.append(full_txt)
        locations.append(loc)
    except:
        logging.exception(f"PROBLEM WITH SOME SHET!-{url}")
        print(url)
        break
```

```python
count = 0
indexes = []
for i, loc in enumerate(locations):
    if re.search('(\d){1,2} \w+ \d{4}', loc):
        count += 1
        indexes.append(i)
```
