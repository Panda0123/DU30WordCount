```python
def format_date(dt):
    #format properly
    day = dt[:2]
    month_yr = dt[2:].strip()
    dt = f'{month_yr[:-4].strip()} {day}, {month_yr[-4:]}' 
    return dt
```

```python
dates = []
locations = []
locs = []
for i in df_date.index.tolist():
    url = df_date.loc[i, 'url']
    response = requests.get(url, headers=headers)
    html = BeautifulSoup(response.text, 'lxml')
    container = html.find('div', 'td-post-content')
    p_list = container.findAll('p')
    pattern_index = 0
    span = [0, 1]
    for i in range(len(p_list)):
        loc = p_list[i].text.replace('\n', ' ',).strip()
        if re.search(r'\[(.*?)\]', loc):
            if loc.find('|') != -1:
                pattern_index = i
                locs.append(loc)
                span = re.search(r'\[(.*?)\]', loc).span()
                break
    pattern_str = loc[span[0]:span[1]]
    loc = pattern_str[pattern_str.find('|')+ 1:-1].strip()
    date = pattern_str[1:pattern_str.find('|')].strip()
    
    #some are formated differently
    if re.search('(\d){1,2} \w+ \d{4}', loc):
        loc = pattern_str[1:pattern_str.find('|')].strip()
        date = pattern_str[pattern_str.find('|') + 1:-1].strip()
    
    if bool(re.search('(\d){1,2} \w+ \d{4}', date)):
        date = format_date(date)
        
    if 'Delivered at' in loc:
        loc = loc[loc.find('Delivered at')+12:].strip()
        
    df.loc[i, 'date'] = date
    df.loc[i, 'locations'] = loc
    dates.append(date)
    locations.append(loc)
```

# Identify which transcript in df2_same_date is already in df

>I'm gonna create a function that basically check if the first nth words are the same. If they are then they are probably the same transcripts.

```python
def is_same(trans1, trans2, num_words=-1):

    trans1_lt_temp = trans1.replace('\n', ' ').split(" ")
    trans1_lt = []
    
    #remove extra spaces
    for i in trans1_lt_temp:
        if i != '':
            trans1_lt.append(i)
            
    trans2_lt_temp = trans2.replace('\n', ' ').split(" ")
    trans2_lt = []
    for i in trans2_lt_temp:
        if i != '':
            trans2_lt.append(i)   
            
    if num_words == -1:
        num_words= min(len(trans1_lt), len(trans2_lt))

    for i in range(num_words):
        if trans1_lt[i] != trans2_lt[i]:
            return False
    
    return True

def is_same_last(trans1, trans2, num_words=-1):
    trans1_lt_temp = trans1.replace('\n', ' ').split(" ")
    trans1_lt = []
    
    #remove extra spaces
    for i in trans1_lt_temp:
        if i != '':
            trans1_lt.append(i)
    
    trans2_lt_temp = trans2.replace('\n', ' ').split(" ")
    trans2_lt = []
    for i in trans2_lt_temp:
        if i != '':
            trans2_lt.append(i)   
        
    
    if num_words == -1:
        num_words= min(len(trans1_lt), len(trans2_lt))
    
    #get only the last num_words
    trans1_lt = trans1_lt[-num_words:]
    trans2_lt = trans2_lt[-num_words:]
    for i in range(num_words):
        if trans1_lt[i] != trans2_lt[i]:
            return False
    
    return True

is_same('wa sup mga paa', 'wa sup    mga paa', 4)
```
