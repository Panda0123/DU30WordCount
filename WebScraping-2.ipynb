```python
def format_date(dt):
    #format properly
    day = dt[:2]
    month_yr = dt[2:].strip()
    dt = f'{month_yr[:-4].strip()} {day}, {month_yr[-4:]}' 
    return dt

def ordinary(html, lnk):
    #scrape title
    heading = html.find('div', class_='page-header').text.strip()
    separator_index = heading.find('–')
    title = heading[separator_index+ 1:].strip()
    container = html.find('table')
    
    #scrape date -some formatter differently
    try:
        dt = container.findAll('td')[2].text.strip()
        if len(dt) > 30:
            dt = container.findAll('td')[1].text.strip()
    except:
        dt = html.find('div', class_='col-xs-12 col-sm-12').findAll('p')[2].text.strip()
        if len(dt) > 30:
            dt = html.find('div', class_='col-xs-12 col-sm-12').findAll('p')[0].findAll('span')[2].text
    
    if dt.find(',') != 1:
        dt = format_date(dt)
    
    #scrape location -some formatted differently
    try:
        location = container.tbody.findAll('tr')[1].text.strip()
    except:
        location = html.find('div', class_='col-xs-12 col-sm-12').findAll('p')[1].text

    #scrape transcript - some are formatted differently
    try:
        full_txt = html.find('tbody').findAll('td')[3].text
        #again some are formatted fucking differently (just like his words very inconsistent)
        if len(full_txt) < 30:
            full_txt = html.find('tbody').findAll('td')[4].text
    except:
        full_txt = ''
        container_paragraph = html.find('div', class_='col-xs-12 col-sm-12')
        paragraphs = container_paragraph.findAll('p')
        for j in range(3, len(paragraphs)):
            full_txt += paragraphs[j].text.strip()
    
    lt_names = ['PRESIDENT RODRIGO DUTERTE:', 'PRESIDENT DUTERTE:', 'PRESIDENT RODRIGO ROA DUTERTE:']

    #check if it is a multi dialogue(conversation) transcript
    if any(nm in full_txt for nm in lt_names):
        full_txt = ''
        isDuterte = True
        container = html.find('tbody').findAll('td')[3]
        for p in container.findAll('p'):
            if p.strong:
                    if any(nm in p.strong.text.strip() for nm in lt_names):
                        isDuterte = True
                    else:
                        isDuterte = False

            if isDuterte:
                full_txt += p.text.strip()

    return title, dt, location, np.nan, full_txt

def multi_dialogue(html, lnk):
    try:
        title = html.find('h3', class_='entry-title').text.strip()
        date = html.find('time').text.strip()
    except:
        print(f'No title:({lnk})')
    
    speech_info = html.findAll('div', class_='speech-info')
    
    #some does not contain event and just location or does not contain both
    if len(speech_info) > 1:
        event = speech_info[0].findAll('span')[1].text.strip()
        location = speech_info[1].findAll('span')[1].text.strip()
    elif len(speech_info) == 1:
        event = np.nan
        location = speech_info[0].findAll('span')[1].text.strip()
    else:
        event = np.nany
        location = np.nan
    
    body = html.find('div', class_='release-content')
    full_txt = ''
    isDuterte = True
    lt_names = ['PRESIDENT RODRIGO DUTERTE:', 'PRESIDENT DUTERTE:', 'PRESIDENT RODRIGO ROA DUTERTE:']

    for p in body.findAll('p'):
        if p.strong:
            if any(nm in p.strong.text.strip() for nm in lt_names):
                isDuterte = True
            else:
                isDuterte = False

        if isDuterte:
            full_txt += p.text.strip()
            
    return title, date, location, event, full_txt

def download_pdf(container_tr, reponse2, date_counter):
    tl_dt = container_tr[i].text.strip()
    comma = tl_dt.rfind(',')
    title = tl_dt[:comma]
    dt = tl_dt[comma+1:].strip()
    dt = dt[:dt.find('|')].strip()

    #format properly
    dt = format_date(dt) #will be used as file name
    date = dt #will be returned for dictionary key
   
    #so check if date already exist
    if dt in date_counter:
        dt = f'{dt}_{date_counter[date] + 1}'
#     for fl in os.listdir('duterte_pdf2'):
#         if dt == fl[:-4].strip():
#             dt = f'{dt}_{date_counter[date] + 1}'
    try:
        with open(f'duterte_pdf2\\{dt}.pdf', 'wb') as f:
            f.write(response2.content)
            return title, date, np.nan, np.nan, np.nan
    except:
        print('Failed download pdf', date)
        
```

```python
import logging
import numpy as np

years = [2016, 2017, 2018]
months = ['january', 'february', 'march', 'april', 'may', 'june', 'july', 
          'august', 'september', 'october', 'november', 'december']

lt_not_loaded = []
yr_month_non_existent = []
lt_404 = []
information = {}
date_counter = {}
for yr in years:
    for month in months:
        url = f'https://pcoo.gov.ph/{month}-{yr}-president-rodrigo-roa-dutertes-speeches/'
        response = requests.get(url=url, headers=headers)
        
        if response.status_code == 200:
            try:
                html_lt_speeches = BeautifulSoup(response.text, 'lxml')
                #this contains list of speeches in that date(yr-month)
                container = html_lt_speeches.find('table')
                container_tr = container.findAll('tr')

                #traverse through the speeches
                for i in range(2, len(container_tr)):
                    a_tags = container_tr[i].findAll('a')
                    #locate the URL 
                    for a_tag in a_tags:
                        if a_tag.text.strip() == 'Transcript':
                            lnk = a_tag['href'].strip()
                                
                    response2 = requests.get(lnk, headers=headers)
                    
                    if response2.status_code == 200:
                        #check if lnk is a url for pdf format of the speech or just page
                        if lnk[-4:] != '.pdf':
                            html_page = BeautifulSoup(response2.text, 'lxml')

                            #some are formatted differently/contain conversation(interviews)
                            try:
                                title, date, location, event, transcript = ordinary(html_page, lnk)
                            except:
#                                 logging.exception(f"An exception was thrown(PROBLEM a-tag)!-{lnk}") 
                                title, date, location, event, transcript = multi_dialogue(html_page, lnk)
                    
                        else:#if pdf
                            title, date, location, event, transcript= download_pdf(container_tr, response2, date_counter)
                            
                            #temporary check
                            if len(date) > 30:
                                print(f'Long Date:dt')
                                print(lnk)
                                print()
                        #if date already exist
                        if date in information.keys():
                            lt_keys = ['title', 'date', 'location', 'event', 'url']
                            tempDict = {'title':title, 'date':date, 'location':location, 'event':event, 'url':lnk}
                            simDict = information[date]
                            
                             #check if same speech 
                            isSame = True
                            for key in lt_keys:
                                #check if nan
                                if type(tempDict[key]) == float and type(simDict[key]) == float:
                                    pass
                                elif tempDict[key] != simDict[key]:
                                    isSame = False
                                    break
                                
                            if isSame:
                                print(f'Already exist speech:{date, title}')
                            else: 
                                #same date but different speech
                                #since date is our key we're going to add a suffix to this which is the counter
                                counter = date_counter[date] + 1 
                                information[date + f'_{counter}'] = {'title':title, 'date':date, 'location':location, 
                                                 'event':event, 'transcript':transcript, 'url':lnk}
                                date_counter[date] = counter
                        else:
                            information[date] = {'title':title, 'date':date, 'location':location, 
                                                 'event':event, 'transcript':transcript, 'url': lnk}
                            date_counter[date] = 1
                          
                           
                    else:
                        lt_404.append(lnk)

            except BaseException:
                logging.exception(f"An exception was thrown (PROBLEM table-tag)!-{url}")
                print(f'Empty Page:{lnk}')
                lt_not_loaded.append(lnk)
        else:
            yr_month_non_existent.append((yr, month))
```
