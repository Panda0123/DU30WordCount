```python
def is_in(txt):
    if any(((txt.lower() in trans) or (trans in txt.lower())) for trans in df.transcript.values.tolist()):
        return True
    else:
        return False

def clean_transcript(single_str):
    lt_names= ['PRESIDENT RODRIGO DUTERTE:', 'PRESIDENT DUTERTE:', 'PRESIDENT RODRIGO ROA DUTERTE:', 'DUTERTE:', 
               'PRES. DUTERTE:', 'PRRD']
    str_cleaned = single_str
    for nm in lt_names:
        str_cleaned = str_cleaned.replace(nm, ' ')
    str_cleaned = str_cleaned.replace('\xa0', ' ')
    str_cleaned = str_cleaned.replace('\x0c', ' ')
    str_cleaned = str_cleaned.replace('\\', ' ')
    str_cleaned = str_cleaned.replace('—', ' ')
    str_cleaned = str_cleaned.replace('-', ' ')
    str_cleaned = str_cleaned.replace('...', ' ')
    str_cleaned = str_cleaned.replace('…', ' ')
    str_cleaned = str_cleaned.replace('END', ' ')
    str_cleaned = re.sub(r"\[(.*?)\]", ' ', str_cleaned) #remove all that is contained within brackets
    str_cleaned = re.sub(r"\((.*?)\)", ' ', str_cleaned) #remove all that is contained within parentheses
    str_cleaned = re.sub(r'\“(.+?)\”', ' ', str_cleaned) #remove all that is in quatation as he probably quoting someone
    str_cleaned = str_cleaned.replace('‘', ' ')
    str_cleaned = [wrd.lower() for wrd in str_cleaned.split(' ') if wrd != '']
    str_cleaned = ' '.join(str_cleaned)

    return str_cleaned
```

```python
def is_same(trans1, trans2, num_words=-1):

    trans1_lt_temp = trans1.replace('\n', ' ').split(" ")
    trans1_lt = []
    
    #remove extra spaces
    for i in trans1_lt_temp:
        if i != '':
            trans1_lt.append(i)
            
    trans2_lt_temp = trans2.replace('\n', ' ').split(" ")
    trans2_lt = []
    for i in trans2_lt_temp:
        if i != '':
            trans2_lt.append(i)   
            
    if num_words == -1:
        num_words= min(len(trans1_lt), len(trans2_lt))

    for i in range(num_words):
        if trans1_lt[i] != trans2_lt[i]:
            return False
    
    return True

def is_same_last(trans1, trans2, num_words=-1):
    trans1_lt_temp = trans1.replace('\n', ' ').split(" ")
    trans1_lt = []
    
    #remove extra spaces
    for i in trans1_lt_temp:
        if i != '':
            trans1_lt.append(i)
    
    trans2_lt_temp = trans2.replace('\n', ' ').split(" ")
    trans2_lt = []
    for i in trans2_lt_temp:
        if i != '':
            trans2_lt.append(i)   
        
    
    if num_words == -1:
        num_words= min(len(trans1_lt), len(trans2_lt))
    
    #get only the last num_words
    trans1_lt = trans1_lt[-num_words:]
    trans2_lt = trans2_lt[-num_words:]
    for i in range(num_words):
        if trans1_lt[i] != trans2_lt[i]:
            return False
    
    return True
```

>Let's create a funciont that check if first nth words is the same to others transcript in df.

```python
def is_same_with_others(ser, nth=-1):
    dictionary = {}
    for i in range(ser.shape[0]):
        trans1 = ser.values[i]
        trans1_lt_temp = trans1.replace('\n', ' ').split(" ")
        trans1_lt = []

        #remove extra spaces
        for wrd in trans1_lt_temp:
            if wrd != '':
                trans1_lt.append(wrd)
        for j in range(i +1, ser.shape[0]-1):
            trans2 = ser.values[j]
            trans2_lt_temp = trans2.replace('\n', ' ').split(" ")
            trans2_lt = []
            for wrd in trans2_lt_temp:
                if wrd != '':
                    trans2_lt.append(wrd)   

            if nth == -1:
                nth= min(len(trans1_lt), len(trans2_lt))
            isSame = True
            for k in range(nth):
                if trans1_lt[k] != trans2_lt[k]:
                    isSame = False
                    break
                
            if isSame:
                if i in dictionary.keys():
                    dictionary[i] = dictionary[i].append(j)
                else:
                    dictionary[i] = [j]
    
    return dictionary
```
